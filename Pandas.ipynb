Pandas DataFrame : is two dimensnal tabular data structure with labeled axes(rows and columns)

```python
#importing pandas
import pandas as pd
```

```python
#create a pandas dataframe
#importing the Boston house price data
from sklearn.datasets import fetch_california_housing
california_dataset = fetch_california_housing()
type(california_dataset)
print(california_dataset)
```

```python
# Pandas DataFrame
california_df = pd.DataFrame(california_dataset.data, columns = california_dataset.feature_names)
california_df.head()
california_df.shape
```

```python
#Imprting the data frome CSV file to a pandas dataframe
diabetes_df = pd.read_csv('pathe of CSV file')
type(diabetes_df)
diabetes_df.head()
diabetes_df.shape
```

```python
# Loading data from a excel file to a pandas dataframe
pd.read_excel('file path')
```

```python
# Exporting a dataframe to a csv file
california_df.to_csv('california.csv')
```

```python
# Exporting the pandas dataframe to excel file
california_df.to_excel('california.xlsx')
```
