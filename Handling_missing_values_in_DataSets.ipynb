```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the datasets to a pandas dataframe
dataset = pd.read_csv('/content/sample_data/melb_data.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
# Missing value
dataset.isnull().sum()
```

```python
# analysis the distribution of data in car column
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(dataset.Car)
```

```python
# replacing the missing values with the median value
dataset['Car'].fillna(dataset['Car'].median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# replacing the missing values with the median value
dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean(),inplace=True)
```

```python
# replacing the missing values with the median value
dataset['CouncilArea'].fillna(dataset['CouncilArea'].mode(),inplace=True)
```

```python
dataset.isnull().sum()
```
