Seaborn is a Data visualization library

```python
#importing the library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some builtin datasets

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Visulizing the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue = 'smoker', style='smoker', size='size')
```

```python
# Setting a Theme for plots
sns.set_theme()
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue ='species', data = iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue ='species', data = iris)
```

```python
# Loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Count Plot
sns.countplot(x='class',data = titanic)
sns.countplot(x='survived', data = titanic)
```

```python
# Bar Chart
sns.barplot(x='sex', y = 'survived',hue='class', data=titanic)
```

```python
# House Price dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house = pd.DataFrame(housing.data, columns = housing.feature_names)
house['PRICE'] = housing.target
house.head()
```

```python
# Distribution Plot
sns.displot(house['PRICE'])
```

```python
# HeatMape
correlation = house.corr()
```

```python
#constructing a Heat Mape
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
