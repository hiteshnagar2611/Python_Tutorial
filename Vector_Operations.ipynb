```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,-3,-6, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Adding two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sum1 = vector_1 + vector_2
print(sum1)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Substraction of Two Vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sub = vector_1 - vector_2
print(sub)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scaler

```python
vector_1 = np.asarray([0,0,2,3])

multi = -2 * vector_1
print(multi)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-4,-6, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product of 2 vectors

```python
a = np.asarray([2,3])
b = np.asarray([4,4])

a_dot_b =np.dot(a,b)
print(a_dot_b)
```

```python
c = np.asarray([24,31,34])
d = np.asarray([42,44,21])

c_dot_d =np.dot(c,d)
print(c_dot_d)
```

Cross product of 2 vectors

```python
a = np.asarray([2,3,2])
b = np.asarray([4,4,5])

a_cross_b = np.cross(a,b)
print(a_cross_b)
```

Projection of 'a' vector on 'v' vector

```python
a = np.asarray([2,5])
v = np.asarray([8,-6])

# magnitude of v vector
magnitude_of_v = np.sqrt(sum(v**2))

projection_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print(projection_of_a_on_v)

```
